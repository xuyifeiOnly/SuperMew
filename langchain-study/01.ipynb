```python
import langchain
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

API_KEY=os.getenv("ARK_API_KEY")
MODEL=os.getenv("MODEL")
BASE_URL=os.getenv("BASE_URL")
```

```python
from langchain.chat_models import init_chat_model
from langchain.agents import create_agent

model=init_chat_model(
    model=MODEL,
    model_provider="openai",
    base_url=BASE_URL,
    api_key=API_KEY
)
```

```python
model.invoke("nihao")
```

```python
from langchain_core.messages import HumanMessage,AIMessage,SystemMessage

message=[
    SystemMessage(content="你是一只小猫猫，擅长棠棠喵")
]
print("多轮对话开始啦🌶：（输入q退出）")
while True:
    user_input=input("🥸 你：")
    if user_input.lower()=='q':
        break
    message.append(HumanMessage(content=user_input))

    respose=model.invoke(message)

    message.append(AIMessage(content=respose.content))

    print(f"🤖 AI:{respose.content}\n")
```

```python
message=[
    SystemMessage(content="你是一只小猫猫，擅长棠棠喵")
]
print("多轮对话开始啦🌶：（输入q退出,这次是流式的）")
while True:
    user_input=input("🥸 你：")
    if user_input.lower()=='q':
        break
    message.append(HumanMessage(content=user_input))

    print("🤖 AI:")

    full_response=""
    for chunk in model.stream(message):
        print(chunk.content,end="",flush=True)
        full_response+=chunk.content

    # respose=model.invoke(message)

    message.append(AIMessage(content=full_response))

    
```

```python
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def get_keywords(word:str)->str:
    """获取关键词,'耄耋','毫猫'和英文"""
    if word=="耄耋":
        return "😼哈！"
    if word=="毫猫":
        return "😽喵！"
    return "🙀没有人类了！"

agent=create_agent(
    model=model,
    tools=[get_keywords],
    system_prompt="你是一个哈基米，听不懂英文话"
)

result=agent.invoke({
    "messages":[{
        "role":"user",
        "content":"你是谁"
    }]
})
print(f"{result['messages'][-1].content}")

print()
```
